```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    'ID': ['P01', 'P02', 'P03', 'P04', 'P05', 'P06', 'P07', 'P08'],
    'Price': [4999, 799, 399, 99, 1299, 249, 699, 109],
    'Sales': [120, 85, 60, 300, 90, 150, 110, 270],
    'Stock': [15, 32, 8, 45, 20, 40, 22, 35],
    'Price_Diff': [100, 50, -21, 4, 200, -10, 30, 9],
    'Cat_Code': [1, 2, 3, 4, 2, 4, 2, 4]
}

df = pd.DataFrame(data)
df
```

```python
X = df[['Sales', 'Stock', 'Price_Diff', 'Cat_Code']]
y = df['Price']

model = LinearRegression()
model.fit(X, y)

df['Predicted_Price'] = model.predict(X)
df
```

```python
mae = mean_absolute_error(y, df['Predicted_Price'])
mse = mean_squared_error(y, df['Predicted_Price'])
rmse = np.sqrt(mse)
mape = mean_absolute_percentage_error(y, df['Predicted_Price'])

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("MAPE:", mape)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='Predicted_Price', data=df)
plt.plot([min(y), max(y)], [min(y), max(y)], color='red')
plt.title('Actual vs Predicted Price')
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Cat_Code', y='Sales', data=df)
plt.title('Sales by Product Category')
plt.xlabel('Category Code')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Cat_Code', y='Price', data=df)
plt.title('Price Distribution per Category')
plt.xlabel('Category Code')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
